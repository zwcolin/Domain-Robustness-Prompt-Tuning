```python
def get_epoch_loss(file):
    re.sub("[^\d\.]", "", "$1,999,888.77")
    epoch, loss = [], []
    with open(file, 'r') as f:
        lines = f.readlines()
        for line in lines:
            if 'learning_rate' in line:
                line = line.split(':')
                loss.append(float(line[1][1:7].replace(',', '').replace(' ', '').replace("'", '')))
                epoch.append(float(line[3][1:5].replace(',', '').replace(' ', '').replace('}', '')))
    return epoch, loss
```

```python
import plotly.graph_objects as go
```

```python
def smooth(x, y, span=4):
    x_new, y_new = [], []
    for i in range(0, len(y), span):
        x_new.append(mean(x[i:i+span]))
        y_new.append(mean(y[i:i+span]))
    return x_new, y_new
```

```python
plt = go.Figure()
for key in d:
    x, y = get_epoch_loss(d[key])
    x, y = smooth(x, y)
#     plt.add_trace(go.Scatter(x=x, y=y, name="spline",
#                     text=["tweak line smoothness<br>with 'smoothing' in line object"],
#                     hoverinfo='text+name',
#                     line_shape='spline'))
    plt.add_trace(go.Scatter(x=x, y=y,name=key,))
plt.update_traces(mode="lines", hovertemplate=None)
plt.update_layout(hovermode="x")
plt.update_xaxes(title_text='Epoch')
plt.update_yaxes(title_text='Loss')
plt.show()
```
