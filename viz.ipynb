```python
import matplotlib.pyplot as plt
import json
```

```python
loss, epoch = [], []
# f = open('prompt_tuning/squad/t5-small/10/2022-01-21-185719/log.txt', "r")
f = open('prompt_tuning/squad/t5-base/10/2022-01-23-123803/log.txt', "r")


for line in f.readlines():
    line = line.strip()
    if 'loss' in line and 'train_runtime' not in line:
        d = json.loads(f"{line}".replace("'", "\""))
        loss.append(d['loss'])
        epoch.append(d['epoch'])
```

```python
loss1, epoch1 = [], []
f = open('prompt_tuning/squad/t5-small/10/2022-01-21-185719/log.txt', "r")

for line in f.readlines():
    line = line.strip()
    if 'loss' in line and 'train_runtime' not in line:
        d = json.loads(f"{line}".replace("'", "\""))
        loss1.append(d['loss'])
        epoch1.append(d['epoch'])
```

```python
ax = plt.subplot()

ax.plot(epoch, loss, label='Training Loss, Base')
ax.plot(epoch1, loss1, label='Training Loss, Small')

plt.xlabel('Epoch')
plt.ylabel('Loss')
ax.legend()
```
